```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## EXPONENTIAL LINEAR UNIT

```python
# specifing the value of alpha 0.9

alpha = 0.9

list1 = []
for i in np.arange(-10,10,0.01):
  if i>=0:
    list1.append(i)
  else:
    list1.append(alpha*(np.exp(i)-1))

plt.plot(np.arange(-10,10,0.01),list1)
plt.xlabel("x")
plt.ylabel("EXPONENTIAL LINEAR UNIT")
plt.title("EXPONENTIAL LINEAR UNIT ACTIVATION FUNCTION")
plt.show()
```

## exponential linear unit derivative

```python
# specifing the value of alpha 0.9

alpha = 0.9

list1 = []
for i in np.arange(-10,10,0.01):
  if i>=0:
    list1.append(1)
  else:
    list1.append(alpha*(np.exp(i)))

plt.plot(np.arange(-10,10,0.01),list1)
plt.xlabel("x")
plt.ylabel("EXPONENTIAL LINEAR UNIT")
plt.title("EXPONENTIAL LINEAR UNIT ACTIVATION FUNCTION DERIVATION ")
plt.show()
```
