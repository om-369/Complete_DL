```python
import numpy as np
import matplotlib.pyplot as plt
```

## RELU ACTIVATION FUNCTION

```python
list1 = []

for i in np.arange(-10,10,0.01):
  list1.append(max(i,0))

plt.plot(np.arange(-10,10,0.01),list1)
plt.xlabel("x")
plt.ylabel("RELU")
plt.title("RELU ACTYIVATION FUNCTION ")
plt.show()
```

## LEAKY RELU ACTIVATION FUNCTION (0.1)

```python
list1 = []

for i in np.arange(-10,10,0.01):
  if i>0:
    list1.append(i)
  else:
    list1.append(0.1*i)


plt.plot(np.arange(-10,10,0.01),list1)
plt.xlabel("x")
plt.ylabel("LEAKY RELU")
plt.title("LEAKY RELU ACTYIVATION FUNCTION ")
plt.show()
```
