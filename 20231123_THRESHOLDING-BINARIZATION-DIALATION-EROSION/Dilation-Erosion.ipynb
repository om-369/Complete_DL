```python
# define our imshow function

def imshow(title = "Image",image = None,size = 7):

  w,h = image.shape[0],image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
"""

1. Dilation -- adds pixel to the boundary of image
2. Erosion -- removes pixels from boundaries of image
3. Opening -- erosion followed by dilation
4. Closing -- dilation followed by erosion

"""
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load the image as grey scale

image = cv2.imread("/content/images/opencv_inv.png",0)
imshow("Original Image",image)


# lets define the kernel size
kernel = np.ones((5,5),np.uint8)

# lets do erosion
erosion = cv2.erode(image,kernel,iterations = 1)
imshow("Erosion Image",erosion)

# lets do dilation
dilation = cv2.dilate(image,kernel,iterations = 1)
imshow("Dilation Image",dilation)

# # lets do dilation
# dilation_2 = cv2.dilate(image,kernel,iterations = 3)
# imshow("Dilation_3 Image",dilation_2)

opening = cv2.morphologyEx(image,cv2.MORPH_OPEN,kernel)
imshow("Opening",opening)

closing = cv2.morphologyEx(image,cv2.MORPH_CLOSE,kernel)
imshow("closing",closing)
```
