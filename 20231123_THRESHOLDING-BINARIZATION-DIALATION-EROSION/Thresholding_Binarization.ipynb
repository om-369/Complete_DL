```python
# define our imshow function

def imshow(title = "Image",image = None,size = 7):

  w,h = image.shape[0],image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Thresholding, Binarization, Adaptive Thresholding

```python
# load the image as grey scale

image = cv2.imread("/content/scan.png",0)
imshow("Original Image",image)

# values below 127 will turn black and values above 127 turn white
ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
imshow("1 threshold binary @ 127",thresh1)

# values below 127 will turn white and values above 127 turn black
ret,thresh2 = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
imshow("2 threshold binary @ 127",thresh2)

# values above 127 are held and remaining are changed
ret,thresh3 = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
imshow("3 threshold TRUNC @ 127",thresh3)

ret,thresh4 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
imshow("4 THRESH_TOZERO @ 127",thresh4)

ret,thresh5 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
imshow("5 THRESH_TOZERO @ 127",thresh5)
```

# ADAPTIVE THRESHOLDING

```python
# load the image as grey scale

image = cv2.imread("/content/scan.png",0)
imshow("Original Image",image)

# values below 127 will turn black and values above 127 turn white
ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
imshow("1 threshold binary @ 127",thresh1)

thresh2 = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,3,5)
imshow("ADAPTIVE THRESHOLDING IMAGE",thresh2)

_,thresh3 = cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow("THRESH_OTSU IMAGE",thresh3)
```
