## Shuffled


```python
import tensorflow as tf
import numpy as np
```

```python
# shuffled on axis = 0

not_shuffled = tf.constant([[10,7],
                            [3,4],
                            [2,5]])
not_shuffled
```

```python
tf.random.shuffle(not_shuffled)
```

```python
tf.random.shuffle(not_shuffled,seed = 42)
```

```python
tf.random.shuffle(not_shuffled,seed = 42)
```

```python
# shuffle but in same order

# set global random seed
tf.random.set_seed(42)

tf.random.shuffle(not_shuffled,seed=42)
```
