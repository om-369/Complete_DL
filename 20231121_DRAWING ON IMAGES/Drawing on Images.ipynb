```python
# lets setup first

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define our imshow function

def imshow(title = "Image",image = None,size = 7):

  w,h = image.shape[0],image.shape[1]
  aspect_ratio = w/h
  plt.figure(figsize = (size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

```python
# create an black image using numpy array

rgb_image = np.zeros((512,512,3),np.uint8)

# lets make gray scale image

gray_image = np.zeros((512,512),np.uint8)
```

```python
imshow("Black canvas -RGB COLOR FORMAT ",rgb_image)
```

```python
imshow("Black canvas -GREY COLOR FORMAT ",gray_image)
```

# Rectangle

```python
# create an black image using numpy array

rgb_image = np.zeros((512,512,3),np.uint8)

cv2.rectangle(rgb_image,(100,100),(250,250),(157,50,127),10)

imshow("Black Canvas with Rectangle ",rgb_image)
```

# Circles

cv2.circle(image,center,radius,color,fill)

```python
# create an black image using numpy array

rgb_image = np.zeros((512,512,3),np.uint8)

cv2.circle(rgb_image,(250,250),100,(15,150,50),5)

imshow("Black Canvas with Circle ",rgb_image)
```

```python
# create an black image using numpy array

rgb_image = np.zeros((512,512,3),np.uint8)

cv2.circle(rgb_image,(250,250),100,(15,150,50),-1)

imshow("Black Canvas with painted Circle inside  ",rgb_image)
```

image: It is the image on which the circle is to be drawn.
center_coordinates: It is the center coordinates of the circle. The coordinates are represented as tuples of two values i.e. (X coordinate value, Y coordinate value).
radius: It is the radius of the circle.
color: It is the color of the borderline of a circle to be drawn. For BGR, we pass a tuple. eg: (255, 0, 0) for blue color.
thickness: It is the thickness of the circle border line in px. Thickness of -1 px will fill the circle shape by the specified color.

# Polygons

cv2.polylines(image,points,closed?,color,thickness)

```python
# Parameters: image: It is the image on which circle is to be drawn.

# pts: Array of polygonal curves.

# npts: Array of polygon vertex counters.

# contours: Number of curves.

# isClosed: Flag indicating whether the drawn polylines are closed or not. If they are closed, the function draws a line from the last vertex of each curve to its first vertex.

# color: It is the color of polyline to be drawn. For BGR, we pass a tuple.

# thickness: It is thickness of the polyline edges.

# Return Value: It returns an image.
```

```python
# create an black image using numpy array

rgb_image = np.zeros((512,512,3),np.uint8)

pts = np.array([[25, 70], [25, 160],
                [110, 200], [200, 160],
                [200, 70], [110, 20]],
               np.int32)

pts = pts.reshape((-1,1,2))

isClosed = True

cv2.polylines(rgb_image,[pts],isClosed,(0,255,255),5)

imshow("Black Canvas with Polygon ",rgb_image)
```
