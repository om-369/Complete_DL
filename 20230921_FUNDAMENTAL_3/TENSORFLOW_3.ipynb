```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.constant([[1,2],[3,4],[5,6]])

y = tf.constant([[7,8],[9,10],[11,12]])

```

```python
x,y
```

## Dot product

```python
tf.tensordot(tf.transpose(x),y,axes = 1)
```

```python
tf.matmul(tf.transpose(x),y)
```

```python
tf.tensordot(x,tf.transpose(y),axes = 1)
```

## Changing the data type

```python

b = tf.constant([1.7,7.1])

c = tf.constant([1,7])

b,c
```

```python
b = tf.cast(b,dtype = tf.float16)
```

```python
b
```

```python
c = tf.cast(c, dtype = tf.int16)
```

```python
c
```

## Getting the absolute values

```python
tf.abs(d)
```

## Finding the mean, min, max, sum

```python
element = tf.constant(np.random.randint(low = 0,high = 100,size = 50))
element
```

```python
tf.reduce_min(element)
```

```python
tf.reduce_max(element)
```

```python
tf.reduce_mean(element)
```

```python
tf.reduce_sum(element)
```

## finding the positional maximum and minimum element

```python
element
```

```python
tf.argmax(element)
```

```python
tf.argmin(element)
```

## squaring, log, square root

```python
element = tf.constant(np.arange(1,10))

element
```

```python
tf.square(element)
```

```python
tf.sqrt(element)
```

```python
tf.sqrt(tf.cast(element,dtype = tf.float32))
```

```python
h = tf.cast(element,dtype = tf.float32)
```

```python
tf.math.log(h)
```
