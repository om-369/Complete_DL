## PlOTTING THE ACTIVATION Function

## SIGMOID

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):

  return 1/(1+np.exp(-x))


values = np.arange(-10,10,0.1)

plt.plot(values,sigmoid(values))
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("SIGMOID ACTIVATION FUNCTION")
plt.show()
```

## Derivative of sigmoid

```python
def dt_sigmoid(x):

  variable_1 =  1/(1+np.exp(-x))
  return variable_1 * (1 - variable_1)


values = np.arange(-10,10,0.1)

plt.plot(values,dt_sigmoid(values))
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("DERIVATIVE SIGMOID ACTIVATION FUNCTION")
plt.show()
```

## tanh Activation Function

```python
def tanh(x):

  return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))


values = np.arange(-10,10,0.1)

plt.plot(values,tanh(values))
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("tanh ACTIVATION FUNCTION")
plt.show()
```

```python
def dt_tanh(x):

  variable_2 = ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
  dt = 1 - (variable_2)**2
  return dt

values = np.arange(-10,10,0.1)

plt.plot(values,dt_tanh(values))
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("DERIVATIVE tanh ACTIVATION FUNCTION")
plt.show()
```
