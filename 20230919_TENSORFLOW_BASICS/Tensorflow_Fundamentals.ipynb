## Creating an random Number

```python
random_tensor = tf.random.Generator.from_seed(42)

random_tensor = random_tensor.normal(shape=(3,2))

random_tensor_1 = tf.random.Generator.from_seed(42)

random_tensor_1 = random_tensor_1.normal(shape = (3,2))
```

```python
random_tensor,random_tensor_1,random_tensor==random_tensor_1
```
