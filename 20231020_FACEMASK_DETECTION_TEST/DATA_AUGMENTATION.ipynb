```python
import matplotlib.pyplot as plt
import keras.utils as image
from PIL import Image
from keras.preprocessing.image import ImageDataGenerator
```

```python
image = Image.open("/content/ranbir_kapoor.jpg")
```

```python
print(image.format)
print(image.size)
print(image.mode)
```

```python
plt.imshow(image)
```

```python
datagen = ImageDataGenerator(
                      rotation_range = 30,
                      shear_range = 0.2,
                      zoom_range = 0.2,
                      horizontal_flip = True,
                      vertical_flip = True,
                      width_shift_range = 0.2,
                      height_shift_range = 0.2
)
```

```python
import numpy as np

numpy_image = np.asarray(image)

numpy_image.shape
```

```python
input_image = numpy_image.reshape(1,1000,1000,3)
```

```python
i = 0

for output in datagen.flow(input_image,batch_size = 1,save_to_dir = "/content/data_augmentation_ranbir"):
  i+=1
  if i==10:
    break
```
